```python
from sklearn.datasets import load_diabetes
```

```python
data=load_diabetes()
data1=pd.DataFrame(data.data,columns=data.feature_names)
data1['Target']=data.target
```

```python
data1
```

```python
data1.describe()
```

```python
X_train1,X_test1,y_train1,y_test1=train_test_split(data.data,data.target,test_size=0.2,random_state=2)
```

```python
from sklearn.metrics import r2_score
```

```python
coefs=[]
r2=[]
for i in [0,0.1,1,10]:
    reg=Lasso(alpha=i)
    reg.fit(X_train1,y_train1)
    coefs.append(reg.coef_.tolist())
    y_pred=reg.predict(X_test1)
    r2.append(r2_score(y_test1,y_pred))
```
